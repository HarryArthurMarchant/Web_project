```python
#getting rs names
import pandas as pd
df = pd.read_csv('GBR_data.tsv', sep = '\t')
#print(df.head())

rs = df['ID'].to_list()
```

```python
#### getting gene names
# https://bioinformatics.stackexchange.com/questions/2503/how-to-get-a-list-of-genes-corresponding-to-the-list-of-snps-rs-ids/2504

import pandas as pd
from Bio import Entrez
Entrez.email = "laavanyat121@gmail.com"

#df = pd.read_csv('part_one_df.csv', sep = '\t')
#df_1 = df[['rs value']]
#df_2 = df_1[:100]
snp = []
gene = []

snp_ = []
gene_ = []

for snp_id in rs[900000:910000]:
    try:
        record = Entrez.read(Entrez.elink(dbfrom="snp", 
                                  id=snp_id.replace('rs',''), 
                                  db="gene"))
        results = record[0]['LinkSetDb'][0]['Link']
        for result in results:
            uid = result['Id']
            handle = Entrez.esummary(db="gene", id=uid)
            uid_record = Entrez.read(handle)
            handle.close()
            uid_summary = uid_record["DocumentSummarySet"]['DocumentSummary'][0]
            gene_name = uid_summary['Name']

            print(snp_id, uid, gene_name)
            snp.append(snp_id)
            gene.append(gene_name)
            rs_gene = list(zip(snp, gene))
            #print(rs_gene)
    except:
        snp_.append(snp_id)
        gene_.append('N/A')
        gene_NA = list(zip(snp_, gene_))
        

```
